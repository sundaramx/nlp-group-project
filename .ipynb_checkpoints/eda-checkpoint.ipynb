```python
# get missing value/null/nan/na values from a column
def get_null(df, col):
    na = 0
    for elem in df[str(col)]:
        if pd.isna(elem) == True:
            na += 1
    return na

def get_null_per(df, col):
    num_null = get_null(df, col)
    return num_null / len(df[col])
```
